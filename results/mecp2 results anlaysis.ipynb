```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
sns.set
%matplotlib inline 
```

```python
data1 = pd.ExcelFile("burst networks draft 4.xlsx")
data2 = pd.ExcelFile("similarity metrics draft 4.xlsx")
data3 = pd.ExcelFile("featureMatrixConcatenated draft 4.xlsx")
data4 = pd.ExcelFile("featMatrixavebursts draft 4.xlsx")
data5 = pd.ExcelFile("ave_firing_rates_NEO_edited.xlsx")
```

```python
bndf = data1.parse()
smdf = data2.parse()
cbdf = data3.parse()
andf = data4.parse()
frdf = data5.parse()
```

```python
selecsbdf = sbdf.filter(['DIV', 'genotype', 'ave_firing_rate', 'burst_duration', 'inburst_firing_rate', 'num_nodes_unthreholded', 'num_edges_unthresholded', 'density'], axis=1)
sns.pairplot(selecsbdf, hue ="genotype")
```

```python
ax = sns.factorplot(x = 'DIV', y = 'ave_firing_rate', col = 'genotype', 
                 data = frdf, capsize=.2, palette = "Set2",size=6, aspect=.75)
ax.set_axis_labels("DIV", "edge similarity")

#ax = sns.boxplot(x = 'DIV', y = 'node_similarity', hue = 'genotype', data = smdf, whis =1.5,  dodge=True, palette="Set2")
```
